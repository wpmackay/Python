```python
#Linear Congruential Generator
    #Pick a multiplier and a modulus m, both large integers.
    #Pick stargin integer between 0 and m (exclusive)
    #Generate sequence xn of real numbers by k <- a*k mod(m), xk=k/m
n=10000
    
def RNG(k, a=427419669081, m=999999999989):
    k=a*k%m
    xk=k/m
    return xk

k=linspace(1,999999999989,n)
x=RNG(k)

hist(x,50)
```

```python
plot(k,x,ms=1)
```

```python
def randu(k, a=2**16+3, m=2**31):
    k=a*k%m
    xk=k/m
    return xk
```

```python
k=linspace(1,2**31,n)
x=randu(k)

hist(x,50)
```

```python
plot(k,x,ms=1)
```
