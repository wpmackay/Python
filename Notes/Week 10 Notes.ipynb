```python
#Linear Congruential Generator
    #Pick a multiplier and a modulus m, both large integers.
    #Pick stargin integer between 0 and m (exclusive)
    #Generate sequence xn of real numbers by k <- a*k mod(m), xk=k/m
n=10000
    
def RNG(k, a=427419669081, m=999999999989):
    k=a*k%m
    xk=k/m
    return xk

k=linspace(1,999999999989,n)
x=RNG(k)

hist(x,50)
```

```python
plot(k,x,'b.',ms=1)
```

```python
#Want to estimate area of the "flower" generated by the function using Monte Carlo integration
    #First, create "bounding box" that contains the function
    #Sample points randomly from inside the box
    #Count number of points inside curve
    #Ratio of points inside the curve to points outside the curve will be approximately ratio of areas.
    
#Use random.rand to generate x,y coordinates and transform them so that they are usable in polar form in the interval -3,3
```

```python
n=10000
x=6*random.rand(n)-3
y=6*random.rand(n)-3
total_area=6**2
theta2=arctan2(y,x)
r2=sqrt(x**2+y**2)
polar(theta,r,'k',linewidth=2)
polar(theta2,r2,'b.',ms=1, alpha=0.5)

rf=2+cos(7*theta2)
inside=(rf-r2)>=0
inside=sum(inside)

area_ratio=float(inside)/n
area_curve=area_ratio*total_area
print "The area of the curve is approximately",area_curve
```

```python
n=100000
num_swings=100
num_hits=29
p=0.29

def batting_trial(num_swings,num_hits,p,n):
    swings=random.rand(n,num_swings)
    hits=swings<p
    hits_per_trial=sum(hits,axis=1)
    x=float(sum(hits_per_trial==num_hits))  
    return x/n
```

```python
prob=batting_trial(num_swings,num_hits,p,n)
print prob
```
