```python
#Stock Price Simulation
    #s(t) is price of stock at time t
    #model assumes underlying growth rate with random component
    #s(t+1)=(1+mu+sigma*epsilon)*s(t)
        #mu=growth rate
        #epsion=random component, normally distributed
        #sigma=volatility=amplitude of random component
        
def stock_sim(s_0,t,mu=0,sigma=0.05,trials=1):
    s=empty((trials,t))
    s[:,0]=s_0*ones(trials)
    for i in xrange(1,t):
        epsilon=random.normal(size=trials)
        s[:,i]=(1+mu+sigma*epsilon)*s[:,i-1]
    return s    
```

```python
s_0=1
mu=0.000
sigma=0.05
t_max=365
trials=10000
t=linspace(0,t_max,t_max)
s=stock_sim(s_0,t_max,mu,sigma,trials)

final_price=s[:,364]
hist(final_price,50)
EV=mean(final_price)
loss=final_price<1
p_loss=float(sum(loss))/trials
print "Expected Value =", EV
print "Probability of Losing Money =", p_loss
```

```python
#Golden Section Search
    #Start with an interval [a,b] which contains a minimum
    #let r=(sqrt(5)-1)/2
    #x1=a+(1-r)(b-a)
    #x2=a+r(b-a)
    #f1,f2=f(x1),f(x2)
    #Continue while (b-a)<tol
    #if f1<f2, a,x2,b=a,x1,x2
        #x1=a+(1-r)(b-a)
        #f1,f2=f(x1),f2
    #else, a,x1,b=x1,x2,b
        #x2=a+r(b-a)
        #f1,f2=f2,f(x2)
def f(x):
    return (x - 1.3)**2 + 5
        
def golden_section(f,a,b,tol=1e-4, max_iters=100):
    count=0
    r=(sqrt(5)-1)/2    
    x1=a+(1-r)*(b-a)
    x2=a+r*(b-a)
    f1,f2=f(x1),f(x2)
    while (a-b)<tol:
        if count>max_iters:
            break
        if f1<f2:
            a,x2,b=a,x1,x2
            x1 =a+(1-r)*(b-a)
            f1,f2=f(x1),f1
        else:
            a,x1,b=x1,x2,b
            x2=a+r*(b-a)
            f1,f2=f2,f(x2)
        count+=1
    return a
```
