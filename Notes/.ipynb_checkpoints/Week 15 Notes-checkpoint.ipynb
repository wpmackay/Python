```python
f=open("hamlet.txt")
```

```python
hamlet=f.read()
f.close()
```

```python
hamlet=hamlet.lower()
```

```python
hamlet=hamlet.replace('.','').replace(',','').replace(';','').replace('[','').replace(']','').replace('-','').replace('?','').replace('!','').replace('*','').replace(':','')
```

```python
words=hamlet.split()
```

```python
wordcount={}
for word in words:
    if word in wordcount:
        wordcount[word]+=1
    else:
        wordcount[word]=1
print wordcount
```

```python
sorted([1,2,5,4,8])
```

```python
def get_count(x):
    return(x[1])

sorted_words=sorted(wordcount.items(),key=get_count,reverse=True)
```

```python
print sorted_words
```

```python
counts=[get_count(pair) for pair in sorted_words]
```

```python
figure(figsize=(6,6))
p=pie(counts)
```

```python
labels=[pair[0] for pair in sorted_words]
```

```python
figure(figsize=(6,6))
p=pie(counts,labels=labels)
```
